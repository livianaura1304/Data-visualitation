### Soal 3 (30 poin)
Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.
_Trend analysis_ dilakukan dengan menggunakan hubungan berikut:
![output%2011.png](https://www.dl.dropboxusercontent.com/s/dxmrm74ksz5l9v2/trend_analysis_fig.png?dl=0)

#### Code

```python
def trend(x):
  if x > -0.5 and x <= 0.5:
    return "Slight or No change"
  elif x > 0.5 and x <= 1:
    return "Slight Positive"
  elif x > -1 and x <= -0.5:
      return "Slight Negative"
  elif x > 1 and x <= 3:
      return "Positive"
  elif x > -3 and x <= -1:
      return "Negative"
  elif x > 3 and x <= 7:
      return "Among top gainers"
  elif x > -7 and x <= -3:
      return "Among top losers"
  elif x > 7:
      return "Bull run"
  elif x <= -7:
      return "Bear drop"
```

```python
#HangSeng
df_HS["Day_Perc_Change"] = df_HS["Adj Close"].pct_change()*100
df_HS.dropna(inplace = True, axis = 0)

df_HS["Trend"]= df_HS["Day_Perc_Change"].apply(lambda x:trend(x))

df_HS['Date'] = pd.to_datetime(df_HS['Date'])
df_HS_2019 = df_HS[df_HS['Date'].dt.year == 2019]

#visualization
df_HS_pie_data = df_HS_2019.groupby('Trend')
pie_label = sorted([i for i in df_HS_2019.loc[:, 'Trend'].unique()])
plt.pie(df_HS_pie_data['Trend'].count(), labels = pie_label,
        autopct = '%1.1f%%', radius = 2)
plt.title('Visualisasi Saham HangSeng 2019', y = 1.5)
plt.show()
```

```python
#Nasdaq
df_NS["Day_Perc_Change"] = df_NS["Adj Close"].pct_change()*100
df_NS.dropna(inplace = True, axis = 0)

df_NS["Trend"]= df_NS["Day_Perc_Change"].apply(lambda x:trend(x))

df_NS['Date'] = pd.to_datetime(df_NS['Date'])
df_NS_2019 = df_NS[df_NS['Date'].dt.year == 2019]

#visualization Nasdaq
df_NS_pie_data = df_NS_2019.groupby('Trend')
pie_label = sorted([i for i in df_NS_2019.loc[:, 'Trend'].unique()])
plt.pie(df_NS_pie_data['Trend'].count(), labels = pie_label,
        autopct = '%1.1f%%', radius = 2)
plt.title('Visualisasi Saham Nasdaq 2019', y = 1.5)
plt.show()
```

```python
#Nikkei
df_NK["Day_Perc_Change"] = df_NK["Adj Close"].pct_change()*100
df_NK.dropna(inplace = True, axis = 0)

df_NK["Trend"]= df_NK["Day_Perc_Change"].apply(lambda x:trend(x))

df_NK['Date'] = pd.to_datetime(df_NK['Date'])
df_NK_2019 = df_NK[df_NK['Date'].dt.year == 2019]

#visualization Nikkei
df_NK_pie_data = df_NK_2019.groupby('Trend')
pie_label = sorted([i for i in df_NK_2019.loc[:, 'Trend'].unique()])
plt.pie(df_NK_pie_data['Trend'].count(), labels = pie_label,
        autopct = '%1.1f%%', radius = 2)
plt.title('Visualisasi Saham Nikkei 2019', y = 1.5)
plt.show()
```
